```python


# Activate the Bokeh extension
hv.extension('bokeh')

# Create a map that displays a point for every city
city_map = city_data_df.hvplot.points(
    'Lng', 'Lat',  # Longitude and Latitude for positioning
    geo=True,  # Enable geographic plotting
    tiles='OSM',  # Use OpenStreetMap tiles
    size='Humidity',  # Size of points based on Humidity
    color='Humidity',  # Color of points based on Humidity
    hover_cols=['City', 'Country', 'Humidity'],  # Columns to show on hover
    frame_width=800,  # Width of the map
    frame_height=600,  # Height of the map
    title='City Humidity Map'
)

# Display the map
hv.output(city_map)
```

### Step 2: Narrow down the `city_data_df` DataFrame to find your ideal weather condition

```python
# Narrow down cities that fit criteria and drop any results with null values
temperature_min = 270
temperature_max = 310
max_wind_speed = 4.5
cloudiness = 0

# Filter the DataFrame based on conditions
ideal_weather_df = city_data_df[
    (city_data_df['Max Temp'] > temperature_min) & 
    (city_data_df['Max Temp'] < temperature_max) & 
    (city_data_df['Wind Speed'] < max_wind_speed) & 
    (city_data_df['Cloudiness'] == cloudiness)
]

# Drop any rows with null values
ideal_weather_df.dropna()

# Display sample data
ideal_weather_df
```

### Step 3: Create a new DataFrame called `hotel_df`.

```python
# Use the Pandas copy function to create DataFrame called hotel_df to store the city, country, coordinates, and humidity
hotel_df = ideal_weather_df[['City', 'Country', 'Lat', 'Lng', 'Humidity']].copy()
 

# Add an empty column, "Hotel Name," to the DataFrame so you can store the hotel found using the Geoapify API
hotel_df['Hotel Name'] = None 

# Display sample data
hotel_df
```

### Step 4: For each city, use the Geoapify API to find the first hotel located within 10,000 metres of your coordinates.

```python
# Set parameters to search for a hotel
radius = 10000
params = {
    'categories': 'accommodation.hotel',
    'limit': 1,  # Only fetch the first hotel
    'apiKey': geoapify_key
}

# Print a message to follow up the hotel search
print("Starting hotel search")

# Iterate through the hotel_df DataFrame
for index, row in hotel_df.iterrows():
    # get latitude, longitude from the DataFrame
    lat = row['Lat']
    lon = row['Lng']

    # Add the current city's latitude and longitude to the params dictionary
    params["filter"] = f"circle:{lon},{lat},{radius}"
    params["bias"] = f"proximity:{lon},{lat}"

    # Set base URL
    base_url = "https://api.geoapify.com/v2/places"

    # Make and API request using the params dictionary
    response = requests.get(base_url, params=params)

    # Convert the API response to JSON format
    name_address = response.json()

    # Grab the first hotel from the results and store the name in the hotel_df DataFrame
    try:
        hotel_df.loc[index, "Hotel Name"] = name_address["features"][0]["properties"]["name"]
    except (KeyError, IndexError):
        # If no hotel is found, set the hotel name as "No hotel found".
        hotel_df.loc[index, "Hotel Name"] = "No hotel found"

    # Log the search results
    print(f"{hotel_df.loc[index, 'City']} - nearest hotel: {hotel_df.loc[index, 'Hotel Name']}")

# Display sample data
hotel_df
```

### Step 5: Add the hotel name and the country as additional information in the hover message for each city in the map.

```python
import hvplot.pandas  # Import hvplot for Pandas
import holoviews as hv

# Activate the Bokeh extension
hv.extension('bokeh')

# Create a map with enhanced hover information
city_map = hotel_df.hvplot.points(
    'Lng', 'Lat',  # Longitude and Latitude for positioning
    geo=True,  # Enable geographic plotting
    tiles='OSM',  # Use OpenStreetMap tiles
    size='Humidity',  # Size of points based on Humidity
    color='Humidity',  # Color of points based on Humidity
    hover_cols=['City', 'Country', 'Hotel Name', 'Humidity'],  # Add 'Hotel Name' and 'Country' to hover
    frame_width=800,  # Width of the map
    frame_height=600,  # Height of the map
    title='City Humidity and Nearest Hotel Map'
)

# Display the map
hvplot.output(city_map)
```
